```python
import pickle
from gaspymongo import make_atoms_from_doc
from ase.visualize import view
from ase.io import read, write
from IPython.display import Image
```

## Visualize Carbon dataset

```python
with open("datasets/carbon_cleaned_2306.pkl", 'rb') as fp:
    docs = pickle.load(fp)
print("The length of carbon data set is {}".format(len(docs)))
atoms_list = [make_atoms_from_doc(n) for n in docs]
write('slab.png', atoms_list[-1] * (2, 2, 1), rotation='10z,-80x')
Image(open('slab.png','rb').read())
```
